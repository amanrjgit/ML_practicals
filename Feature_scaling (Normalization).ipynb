```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("datasets/wine_data.csv",header=None,usecols=[0,1,2])
df.columns = ["class_label","alcohol","malic_acid"]
```

```python
df.head()
```

```python
sns.kdeplot(df.alcohol)
```

```python
sns.kdeplot(df.malic_acid)
```

```python
colors = {1:"red",2:"green",3:"blue"}
sns.scatterplot(df["alcohol"],df["malic_acid"],hue = df["class_label"],palette=colors)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(df[["alcohol","malic_acid"]],df["class_label"],test_size=0.3,random_state=1)
```

```python
x_train.shape,x_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
minmax = MinMaxScaler()
```

```python
x_train_scaled = minmax.fit_transform(x_train)
```

```python
x_train_scaled
```

```python
x_train_scaled = pd.DataFrame(x_train_scaled,columns=x_train.columns)
```

```python
sns.kdeplot(x_train["alcohol"])
```

```python
sns.kdeplot(x_train_scaled["alcohol"])
```

```python
x_train_scaled.describe()
```

```python
x_test_scaled = minmax.fit_transform(x_test)
x_test_scaled = pd.DataFrame(x_test_scaled,columns=x_test.columns)
```

```python
x_test_scaled.describe()
```
